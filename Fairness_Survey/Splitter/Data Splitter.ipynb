```python
Data= pd.read_excel(r"/content/gdrive/MyDrive/Colab_Notebooks/OKB/Data/Data.xlsx")


Data= Data.replace({'-2+0.85mm':1, '-4+2mm':2, '-6+4mm':3, '-8+6mm':4, '-9.5+8mm':5,
       '-13.2+9.5mm':6, '-16.2+13.2mm':7, '-19.2+16mm':8, '-26+19.2mm':9})

```

```python
P1 = r"/content/gdrive/MyDrive/Colab_Notebooks/OKB/Data/Train"
P2 = r"/content/gdrive/MyDrive/Colab_Notebooks/OKB/Data/Test"


kf= RepeatedKFold(n_splits= 5, n_repeats=5, random_state= 73838).split(Data )

for iteration, data in enumerate(kf):
    Tidx=data[0]
    Testdx= data[1]
    Fold, TrainSet,TestSet = iteration, Data.iloc[Tidx], Data.iloc[Testdx ]
    # Trainer= Data.iloc[Tidx]
    # print(Trainer )
    TrainOuput= os.path.join(P1 ,("Train"+ str(Fold + 1 )+ ".csv"))
    TrainSet .to_csv(TrainOuput, index= False )

    TestOutput= os.path.join(P2 ,("Test"+ str(Fold +1 )+ ".csv"))
    TestSet .to_csv(TestOutput , index= False )

# counter=
# while counter<6:
#     for Train, Test in kf.split(Data ):
#         TrainSet= Data.iloc[Train]
#         TestSet= Data.iloc[Test]
#         TrainOuput= os.path.join(P1 ,("Train"+ str(counter )+ ".csv"))
#         TrainSet .to_csv(TrainOuput, index= False )
#     
#         TestOutput= os.path.join(P2 ,("Test"+ str(counter )+ ".csv"))
#         TestSet .to_csv(TestOutput , index= False )
#     counter +=1
    
```
